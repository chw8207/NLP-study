### 정규 표현식 실습

```python
import re
```

```python
# .기호
r = re.compile('a.c')
```

```python
r.search('kkk')  # 아무것도 출력되지 않음
```

```python
r.search('abc')
```

```python
r.search('a!c')
```

```python
# ?기호
r = re.compile('ab?c')
```

```python
r.search('abbc')  # 아무것도 출력되지 않음
```

```python
r.search('abc')
```

```python
r.search('ac')
```

```python
# *기호
r = re.compile('ab*c')
```

```python
r.search('a')    # 아무것도 출력되지 않음
```

```python
r.search('ac')
```

```python
r.search('abc')
```

```python
r.search('abbbc')
```

```python
# +기호 : 앞의 문자가 최소 1개여야 함
r = re.compile('ab+c')
```

```python
r.search('ac')   # 아무것도 출력되지 않음
```

```python
r.search('abc')
```

```python
r.search('abbbbc')
```

```python
# ^기호 : 시작 문자열 지정
r = re.compile('^ab')
```

```python
# 아무것도 출력되지 않음
r.search('bbc')
r.search('zab')
```

```python
r.search('abz')
```

```python
# {숫자}기호 : 해당 문자를 숫자만큼 반복
r = re.compile('ab{2}c')
```

```python
# 아무것도 출력되지 않음.
r.search('ac')
r.search('abc')
r.search('abbbbc')
```

```python
r.search('abbc')
```

```python
# {숫자1, 숫자2} 기호 : 숫자1이상 숫자2이하 반복
r = re.compile('ab{2,8}c')
```

```python
# 아무것도 출력되지 않음
r.search('ac')
r.search('abc')
r.search('abbbbbbbbbc')
```

```python
r.search('abbc')
```

```python
r.search('abbbbbbbbc')
```

```python
# {숫자,}기호 : 숫자 이상만큼 반복
# {0,} = *
# {1,} = +
r = re.compile('a{2,}bc')
```

```python
# 아무것도 출력되지 않음
r.search('bc')
r.search('aa')
```

```python
r.search('aabc')
```

```python
r.search('aaaaaaaabc')
```

```python
# []기호 : []안의 문자와 매치
# [a-zA-Z] : 알파벳 전부
# [0-9] : 숫자 전부
r = re.compile('[abc]') #[a-c]와 같음 
```

```python
# 아무것도 출력되지 않음
r.search('zzz')
```

```python
r.search('a')
```

```python
r.search('aaaaaaa')
```

```python
r.search('baac')
```

```python
# 알파벳 소문자 범위 지정
r = re.compile('[a-z]')
```

```python
# 아무것도 출력되지 않음
r.search('AAA')
r.search('111')
```

```python
r.search('aBc')
```

```python
# [^문자]기호 : 문자 제외하고 매치시키기
r = re.compile('[^abc]')
```

```python
# 아무것도 출력되지 않음
r.search('a')
r.search('ab')
r.search('b')
```

```python
r.search('d')
```

```python
r.search('1')
```

### 정규 표현식 모듈 함수 실습

```python
# search() : 정규 표현식 전체에 대하여 문자열 매치여부 확인
# match() : 문자열의 첫 부분부터 정규 표현식과 매치하는지 확인
r = re.compile('ab.')
```

```python
r.match('kkkabc')
```

```python
r.search('kkkabc')
```

```python
# re.split()
# 입력된 정규 표현식을 기준으로 분리
# 공백 기준으로 분리해보기
text = '사과 딸기 수박 메론 바나나'
re.split(' ', text)
```

```python
# 줄바꿈 기준 분리
text = """사과
딸기
수박
메론
바나나"""

re.split('\n', text)
```

```python
# +를 기준으로 분리
text = '사과+딸기+수박+메론+바나나'
re.split('\+', text)
```

```python
# re.findall() : 정규 표현식과 매치되는 모든 문자열 리스트로 리턴
# 매치되는 문자열 없을 땐 빈 리스트
text = """이름 : 김철수
전화번호 : 010 - 1234 - 1234
나이 : 30
성별 : 남"""

re.findall('\d+', text)
```

```python
re.findall('\d+', '문자열입니다')
```

```python
# re.sub()
# 정규 표현식 패턴과 일치하는 문자열을 찾아 다른 문자열로 대체할 수 있음
# 특수 문자를 공백으로 바꿀 때
text = """Regular expression : A regular expression, regex or regexp[1] 
(sometimes called a rational expression)[2][3] is, 
in theoretical computer science and formal language theory, 
a sequence of characters that define a search pattern."""
```

```python
preprocessed_text = re.sub('[^a-zA-Z]', ' ', text)
print(preprocessed_text)
```

### 정규 표현식 텍스트 전처리 실습

```python
text = """100 John    PROF
101 James   STUD
102 Mac   STUD"""
```

```python
# \s+ : 최소 1개 이상의 공백 패턴 찾기
# 공백을 기준으로 분리
re.split('\s+', text)
```

```python
# 숫자만 뽑기
re.findall('\d+', text)
```

```python
# 대문자 행 가져오기
re.findall('[A-Z]', text)
```

```python
# 대문자가 연속으로 네 번 등장
re.findall('[A-Z]{4}', text)
```

```python
# 대문자로 구성된 문자열 갖고오기
re.findall('[A-Z][a-z]+', text)
```
